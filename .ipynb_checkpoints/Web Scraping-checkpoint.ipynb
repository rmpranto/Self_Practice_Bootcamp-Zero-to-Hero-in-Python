# CLASS

```python
# First get the request
result = requests.get('https://en.wikipedia.org/wiki/Grace_Hopper')
```

```python
# Create a soup from request
soup = bs4.BeautifulSoup(result.text,"lxml")
```

```python
soup
```

```python
# note depending on your IP Address, 
# this class may be called something different
first_item=soup.select(".mw-headline")[0]
```

```python
first_item
```

```python
for item in soup.select(".mw-headline"):
    print(item.text)
```
